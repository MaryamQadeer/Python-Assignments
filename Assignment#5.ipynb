```python
""" Question # 01 """
print("----------------------Factorial Calculator-----------------------")
def fact_calc(num):
  fact = 1
  for count in range(num):
    fact *= (count+1)
  print("Factorial of the number" ,num, "is:" , fact)
fact_calc(4)

""" Question # 02 """
print("--------------------Upper Case and Lower Case--------------------")
def upper_lower_case(word):
  upper = 0
  lower = 0
  for character in range(len(word)):
    if word[character].isupper():
      upper += 1
    elif word[character].islower():
      lower += 1
    else:
      continue
  print("Upper Case Count in word***",word,"*** is:",upper)
  print("Lower Case Count in word***",word,"*** is:",lower)
upper_lower_case("This Is to CHeck uPPer loWer cASe")

""" Question # 03 """
print("------------------------Even Numbers-----------------------------")
list_numbers = [2,6,4,5,53,22,3,8,9,5,77]
def even_number(list_numbers):
  for count in range(len(list_numbers)):
    if list_numbers[count]%2 == 0:
      print("The number" ,list_numbers[count], "is Even.")
    else:
      print("The number" ,list_numbers[count], "is not Even.")
even_number(list_numbers)

""" Question # 04 """
print("----------------------------Palindrome---------------------------")
def palindrome(word):
  rev_string = word[len(word)::-1]
  if word == rev_string:
    print("The word",word,"is Palindrome.")
  else:
    print("The word",word,"is not Palindrome.")
palindrome("madam")
palindrome("turkey")

""" Question # 05 """
print("------------------------Prime Number Teller-----------------------")
def prime_calculator(number):
    prime_number = "true"
    for count in range(number-1):
        if number%(count+1) == 0:
            if (count+1) != 1:
                print("The Number ", (count+1) , "doesnot make",number,"Prime Number.")
                prime_number = "false"
    if prime_number == "true":
        print("The Number",number,"is a Prime Number.")
prime_calculator(9)
prime_calculator(19)
prime_calculator(6)
prime_calculator(7)

""" Question # 06 """
print("--------------------------Shopping List---------------------------")
def shopping_list(*items):
    for item in items:
        print(item)
shopping_list("Milk","Eggs","Butter")
```
